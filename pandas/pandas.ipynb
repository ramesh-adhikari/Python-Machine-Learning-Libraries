# Pandas
Pandas is a Python library used for working with data sets. It has functions for analyzing, cleaning, exploring, and manipulating data. The name "Pandas" has a reference to both "Panel Data", and "Python Data Analysis" and was created by Wes McKinney in 2008.

In a nutshell, Pandas is like excel for Python, with tables (which in pandas are called DataFrames), rows and columns (which in pandas are called Series), and many functionalities that make it an awesome library for processing and data inspection and manipulation.

```python
import pandas as pd
df = pd.read_csv("data/train.csv")
```

#### Check dataframe
check out the first n rows of our dataframe using the head method. There is also a tail method to look at the last n. By default if no n is given to these methods they return the first 5 or last 5 instances.

```python
df.head()
# notice how no n is given, to get the first 10 items it would be:
# df.head(10)
```

#### info:
the info method returns the number of rows in the dataframe, the number of columns, the name of each column of the dataframe along with the number of non-null values of such column, and the data type of each column.

```python
#Titanic DataFrame
df.info()
```

#### describe: 
the describe method returns some useful statistics about the numeric data in the dataframe, like the mean, standard deviation, maximum and minimum values, and some percentiles.

```python
#Titanic DataFrame
df.describe()
```

#### include parameter
to get information about the non numerical attributes of our dataframe, however, often the statistics calculated for these attributes are not very useful. These attributes are stored as an object data type.

```python
#Titanic DataFrame 
include_list = ['object', 'float64', 'int64']
df.describe(include = include_list)
```

## Apply, Map, and ApplyMap

###### map:
This function allows us to map the values of a pandas Series, to a set of different values. It is mostly used to convert values from a column of a pandas dataframe to some other kind of value. It is very useful for example when we want to convert a categorical value to a boolean one. Lets see how this can be done. For this we will take the Age column of our dataframe and create a new column called old_or_not that has a value of 1 if the age is above a certain threshold and 0 if not. (Sorry if I offend anybody with the threshold :O )

```python
merged_df = df
```

```python
#using the map method on the 'Age' column  
df.map(lambda age: 1 if age >=45 else 0)
```
